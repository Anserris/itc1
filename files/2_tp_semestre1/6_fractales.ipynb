```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def segment(p,q):
    x1,y1=p
    x2,y2=q
    plt.plot((x1,x2),(y1,y2))
segment((3, 7), (7, 5))
```

```python
x,y=[],[]
for i in range(1000):
    x.append(i*0.01)
    y.append(x[i]+np.cos(x[i]))
plt.axis("equal")# ajuste le tracé
plt.plot(x,y);
```

```python
import numpy as np
def rotation(r,d,a):
    x1,y1=r
    x2=x1+d*(np.cos(a))
    y2=y1+d*(np.sin(a))
    plt.plot((x1,x2),(y1,y2))
    plt.axis("equal")
    return (x2,y2)
rotation((0,0),4,np.pi/2) #tracé du point de coordonnées (0,4)
```

```python
segment((0,0),(0,4))
```

```python
def arbre(r, d, a, n):
    if n == 0: # cas de base : on arrête les appels récursifs
        return
    p=rotation(r,d,a)
    segment(r,p) # dessiner le tronc
    arbre(p,d/1.2,a-np.pi/6,n-1) # dessiner le sous-arbre droit
    arbre(p,d/1.2,a+np.pi/6,n-1) # dessiner le sous-arbre gauche
    plt.axis("equal")
    plt.axis("off")# enlève les axes
arbre((0,0),5,np.pi/2,8)
```
